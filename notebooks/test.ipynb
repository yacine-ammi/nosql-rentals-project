# 1: Imports and Initial Setup

```python
import pandas as pd
import numpy as np
import os
import ast

# Set pandas to display more columns so we can see more of the data
pd.set_option('display.max_columns', 100)
```

# Load the Data

```python
# Define the path relative to the notebook's location
input_csv_path = r'D:\Study\Msc DE\NOSQL\nosql-rentals-project\data\listings_Paris.csv'

# Load the dataframe
df = pd.read_csv(input_csv_path)

# Display the first few rows to see what we're working with
df.head()
```

# Select Relevant Columns

```python
# Create a copy to work on, this is a good practice
df_clean = df.copy()

# List of columns we decided to keep based on our schema plan
relevant_columns = [
    'id', 'last_scraped', 'name', 'description', 'property_type', 'room_type',
    'accommodates', 'bathrooms_text', 'bedrooms', 'beds', 'amenities', 'price',
    'minimum_nights', 'maximum_nights', 'instant_bookable', 'host_id', 'host_name',
    'host_since', 'host_location', 'host_about', 'host_is_superhost',
    'host_listings_count', 'host_has_profile_pic', 'host_identity_verified',
    'neighbourhood_cleansed', 'latitude', 'longitude', 'number_of_reviews',
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
    'review_scores_value', 'first_review', 'last_review', 'has_availability'
]

df_clean = df_clean[relevant_columns]

# Check the new shape and columns
print(f"Shape of the new dataframe: {df_clean.shape}")
df_clean.head(5)
```

# Clean the 'price' column

```python
# Check the data type and a few values before cleaning
print("Before cleaning:", df_clean['price'].dtype)
print(df_clean['price'].head(5).to_list())

# Remove '$' and ',' and convert to numeric (float)
df_clean['price'] = df_clean['price'].str.replace(r'[$,]', '', regex=True).astype(float)

# Check the result
print("\nAfter cleaning:", df_clean['price'].dtype)
print(df_clean['price'].head(5).to_list())
```

# Convert Boolean columns ('t'/'f')

```python
bool_cols = ['instant_bookable', 'host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'has_availability']

# Check a sample before
print("Before:", df_clean['instant_bookable'].unique())

bool_map = {'t': True, 'f': False}
for col in bool_cols:
    df_clean[col] = df_clean[col].map(bool_map).astype(bool)

# Check a sample after
print("After:", df_clean['instant_bookable'].unique())
print("Data type:", df_clean['instant_bookable'].dtype)
```

# Convert Date Columns

```python
date_cols = ['last_scraped', 'host_since', 'first_review', 'last_review']

# Check dtypes before
print("Before:", df_clean[date_cols].dtypes)

for col in date_cols:
    df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')

# Check dtypes after
print("\nAfter:\n", df_clean[date_cols].dtypes)
```

# Handle Missing Numeric and Text Values

```python
# First, let's see which columns have missing values
print("Missing values before handling:\n", df_clean.isnull().sum())
```

```python
# Now, let's fill them in
review_score_cols = [
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
    'review_scores_value'
]
df_clean[review_score_cols] = df_clean[review_score_cols].fillna(0)

# We decided 1 is a sensible default for beds/bedrooms
df_clean['bedrooms'] = df_clean['bedrooms'].fillna(1)
df_clean['beds'] = df_clean['beds'].fillna(1)

# And empty strings for text
text_cols = ['description', 'host_about', 'host_location']
df_clean[text_cols] = df_clean[text_cols].fillna('')

# Check remaining nulls (should be very few now)
print("\nMissing values after handling:\n", df_clean.isnull().sum())
```

# Clean 'bathrooms_text'

```python
# Look at some examples of the messy data
print(df_clean['bathrooms_text'].unique()[:10])

# Use regex to extract the number, then fill any misses with 0
df_clean['bathrooms'] = df_clean['bathrooms_text'].str.extract(r'(\d+\.?\d*)').astype(float)
df_clean['bathrooms'] = df_clean['bathrooms'].fillna(0)
df_clean = df_clean.drop(columns=['bathrooms_text'])

# Check our new 'bathrooms' column
print("\nNew bathrooms column values:\n", df_clean['bathrooms'].unique())
```

# Parse 'amenities' string into a list

```python
# Look at a raw amenities value
print("Raw value:", df_clean['amenities'].iloc[0])
print("Type:", type(df_clean['amenities'].iloc[0]))

# This function safely parses the string
def parse_string_list(s):
    try:
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        return []

df_clean['amenities'] = df_clean['amenities'].apply(parse_string_list)

# Look at the parsed value
print("\nParsed value:", df_clean['amenities'].iloc[0])
print("Type:", type(df_clean['amenities'].iloc[0]))
```

# Final Check 

```python
# Final check on data types and nulls
df_clean.info()

# Define the output path
output_csv_path = '../data/cleaned_listings.csv'

# Save the cleaned dataframe
df_clean.to_csv(output_csv_path, index=False)

print(f"\nSuccessfully saved cleaned data to {output_csv_path}")
```
